```python
from konlpy.tag import Kkma # 꼬꼬마 모듈 사용
kkma = Kkma()
my_text = '한국어 분석을 시작합니다 재미있어요~'
print(kkma.sentences(my_text)) # 문장 분석
print(kkma.nouns(my_text)) # 명사 분석
print(kkma.pos(my_text)) # 형태소 분석


```

```python
from konlpy.tag import Hannanum # 한나눔 모듈 사용
hnn = Hannanum()
my_text = '한국어 분석을 시작합니다 재미있어요~'
print(hnn.nouns(my_text))
print(hnn.morphs(my_text))
print(hnn.pos(my_text))

```

```python
from konlpy.tag import Okt # Okt 모듈 사용 : 단순히 워드클라우드로 사용하고 싶다면 Okt 모듈이 가장 '적절'
okt = Okt()
my_text = '한국어 분석을 시작합니다 재미있어요~'
print(okt.nouns(my_text))
print(okt.morphs(my_text))
print(okt.pos(my_text))
```

```python
# 형태소 분류하고 확인해보기
from konlpy.tag import Okt

ok_twitter = Okt()
sentences_tag = []
bucket_list = []

for sentence in reply_text : 
  morph = ok_twitter.pos(sentence)
  sentences_tag.append(morph)
  print(morph)   # 분석된 1건 결과 확인

for one_sentence in sentences_tag : 
  for word, tag in one_sentence : 
    if tag in ['Noun', 'Adjective'] :
      bucket_list.append(word)
  
print(bucket_list)
# print(sentences_tag,sep='\n')
# print(len(sentences_tag))
```

```python
from konlpy.tag import Okt
from wordcloud import WordCloud
from PIL import Image
import numpy as np


with open("naver_moive_Hugh Jackman.txt",'r',encoding = 'utf-8') as f :
  reply_text = f.readlines()

ok_twitter = Okt()
text = ''
for sentence in reply_text : # 댓글 한 줄씩 처리
  for noun in ok_twitter.nouns(sentence) : # 명사 분리
    if noun in ['영화', '도','번','이번','저','이','것','때','말','볼','그','또','애','편이','그냥','부분','좀','더','편'] :
      pass
    else:  
      text+=noun+' ' #출력할 명사만 뽑아서 저장, 공백으로 분리
print(text)

mask_image = np.array(Image.open('AppleMask.jpg'))
wc = WordCloud(font_path = 'Junicode-Bold.ttf',
               background_color = 'white',
               max_words = 100,
               max_font_size = 80,
               min_font_size = 10,
               mask = mask_image,
               colormap='hsv').generate(text)
wc.to_file('wc_result.png')
```
